```python
import os
import json
import glob

# 🧠 Base directories
base_dir = "/content/Calligrapher_assets"
os.makedirs(base_dir, exist_ok=True)

# 📂 Output directories
dirs = {
    "data_dir": os.path.join(base_dir, "Calligrapher_bench_testing"),
    "cli_save_dir": os.path.join(base_dir, "cli_results"),
    "gradio_save_dir": os.path.join(base_dir, "gradio_results"),
    "gradio_temp_dir": os.path.join(base_dir, "gradio_tmp")
}

# Create output directories
for d in dirs.values():
    os.makedirs(d, exist_ok=True)

# 🔍 Helper function to find latest snapshot folder
def find_snapshot(model_name):
    path_pattern = f"/root/.cache/huggingface/hub/models--{model_name}/snapshots/*"
    matches = glob.glob(path_pattern)
    if not matches:
        raise FileNotFoundError(f"No snapshot found for {model_name}")
    # Use the first snapshot (latest)
    return matches[0]

# 🧱 Detect model paths
path_dict = {
    **dirs,  # merge output dirs
    "base_model_path": find_snapshot("black-forest-labs--FLUX.1-Fill-dev"),
    "image_encoder_path": find_snapshot("google--siglip-so400m-patch14-384"),
    "calligrapher_path": os.path.join(find_snapshot("Calligrapher2025--Calligrapher"), "calligrapher.bin")
}

# 💾 Save JSON
json_path = os.path.join(os.getcwd(), "path_dict.json")
with open(json_path, "w") as f:
    json.dump(path_dict, f, indent=4)

print("✅ path_dict.json created at:", json_path)
print("\n📄 Contents:")
print(json.dumps(path_dict, indent=4))
```

```python
"""
calligrapher_ngrok_server_cli_bridge.py

- Reuses your repo's CLI scripts (infer_calligrapher_self_custom.py / infer_calligrapher_cross_custom.py)
- Exposes a FastAPI endpoint /generate and /sequential via ngrok
- Accepts file uploads OR image/mask/reference URLs
- Returns Base64-encoded PNG output or an error

USAGE
1) Ensure you run this from the root of your Calligrapher repo (so the CLI scripts are importable/runable).
2) Install dependencies:
   pip install fastapi uvicorn pyngrok requests python-multipart pillow
3) Run:
   python calligrapher_ngrok_server_cli_bridge.py
"""

import os
import shutil
import io
import json
import tempfile
import base64
import subprocess
import time
from pathlib import Path
from typing import Optional, List

import requests
from fastapi import FastAPI, Form, UploadFile, File
from fastapi.responses import JSONResponse
from pyngrok import ngrok
import uvicorn
from PIL import Image

app = FastAPI()

# -------------------------
# Config: adjust if needed
# -------------------------
# Path to your repo scripts (relative or absolute). If running from repo root, these should exist.
SELF_SCRIPT = "infer_calligrapher_self_custom.py"
CROSS_SCRIPT = "infer_calligrapher_cross_custom.py"

# Output directory where the CLI scripts write results
DEFAULT_OUTPUT_DIR = "cli_results"  # ensure this matches path_dict.json's cli_save_dir if used

# Timeouts for subprocesses (seconds)
SCRIPT_TIMEOUT = 600  # 10 minutes; increase if needed for longer runs

# -------------------------
# Utility helpers
# -------------------------
def download_to_file(url: str, dest_path: str) -> bool:
    """Download URL to dest_path; return True on success."""
    try:
        resp = requests.get(url, timeout=30)
        resp.raise_for_status()
        with open(dest_path, "wb") as f:
            f.write(resp.content)
        return True
    except Exception as e:
        print(f"[download_to_file] error downloading {url}: {e}")
        return False

def find_latest_result(out_dir: str, prefix: str = "result") -> Optional[str]:
    """Find the most recent file in out_dir that starts with prefix."""
    if not os.path.isdir(out_dir):
        return None
    files = [os.path.join(out_dir, f) for f in os.listdir(out_dir) if f.startswith(prefix)]
    if not files:
        return None
    files.sort(key=os.path.getmtime, reverse=True)
    return files[0]

def run_cli_script(cmd: List[str], cwd: Optional[str] = None, timeout: int = SCRIPT_TIMEOUT) -> (int, str, str):
    """
    Run command list as subprocess, return (returncode, stdout, stderr).
    """
    try:
        proc = subprocess.run(cmd, cwd=cwd, capture_output=True, text=True, timeout=timeout)
        return proc.returncode, proc.stdout, proc.stderr
    except subprocess.TimeoutExpired as e:
        return -1, "", f"timeout: {e}"
    except Exception as e:
        return -1, "", str(e)

# -------------------------
# Main endpoint: /generate
# -------------------------
@app.post("/generate")
async def generate(
    # You can pass either URLs or uploaded files.
    image_url: Optional[str] = Form(None),
    mask_url: Optional[str] = Form(None),
    reference_url: Optional[str] = Form(None),  # optional for cross customization
    prompt: str = Form(...),
    # file uploads (multipart)
    image_file: UploadFile = File(None),
    mask_file: UploadFile = File(None),
    reference_file: UploadFile = File(None),
    # generation parameters
    num_inference_steps: int = Form(50),
    guidance_scale: float = Form(1.0),
    use_context: bool = Form(True),  # forwarded to CLI if applicable
):
    """
    Generate with Calligrapher via existing CLI scripts.
    - If a reference image is provided (file or URL), cross-customization script is used.
    - Otherwise, self-customization script is used.
    Returns: {"image_base64": "..."} on success.
    """
    t0 = time.time()
    tmpdir = tempfile.mkdtemp(prefix="calligrapher_api_")
    print(f"[generate] tmpdir: {tmpdir}")

    try:
        # Save input image
        source_path = os.path.join(tmpdir, "source.png")
        if image_file:
            content = await image_file.read()
            with open(source_path, "wb") as f:
                f.write(content)
        elif image_url:
            if not download_to_file(image_url, source_path):
                return JSONResponse({"error": "Failed to download image_url"}, status_code=400)
        else:
            return JSONResponse({"error": "No input image provided (file or image_url)."}, status_code=400)

        # Save mask
        mask_path = os.path.join(tmpdir, "mask.png")
        if mask_file:
            content = await mask_file.read()
            with open(mask_path, "wb") as f:
                f.write(content)
        elif mask_url:
            if not download_to_file(mask_url, mask_path):
                return JSONResponse({"error": "Failed to download mask_url"}, status_code=400)
        else:
            return JSONResponse({"error": "No mask provided (file or mask_url)."}, status_code=400)

        # Save optional reference image (for cross)
        use_cross = False
        ref_path = None
        if reference_file:
            ref_path = os.path.join(tmpdir, "reference.png")
            content = await reference_file.read()
            with open(ref_path, "wb") as f:
                f.write(content)
            use_cross = True
        elif reference_url:
            ref_path = os.path.join(tmpdir, "reference.png")
            if not download_to_file(reference_url, ref_path):
                return JSONResponse({"error": "Failed to download reference_url"}, status_code=400)
            use_cross = True

        # Decide which CLI script to run
        if use_cross:
            script = CROSS_SCRIPT
        else:
            script = SELF_SCRIPT

        # Ensure script exists
        if not os.path.isfile(script):
            return JSONResponse({"error": f"Required script not found: {script}. Run this from repo root or set script path."}, status_code=500)

        # Prepare output dir for this run
        run_out_dir = os.path.abspath(os.path.join(DEFAULT_OUTPUT_DIR, f"run_{int(time.time())}"))
        os.makedirs(run_out_dir, exist_ok=True)

        # Build CLI command. NOTE: adapt flags if your scripts use different arg names.
        # This tries to be conservative: many inference scripts accept --source, --mask, --reference, --out_dir, --prompt, etc.
        cmd = ["python", script,
               "--source", source_path,
               "--mask", mask_path,
               "--prompt", prompt,
               "--out_dir", run_out_dir,
               "--num_inference_steps", str(num_inference_steps),
               "--cfg_scale", str(guidance_scale),
               "--use_context", str(int(use_context)),
               ]

        if use_cross:
            cmd += ["--reference", ref_path]

        print(f"[generate] Running CLI: {' '.join(cmd)}")
        rc, stdout, stderr = run_cli_script(cmd, cwd=os.getcwd())
        print(f"[generate] rc={rc}\nSTDOUT:\n{stdout}\nSTDERR:\n{stderr}")

        if rc != 0:
            return JSONResponse({"error": "Inference script failed", "stdout": stdout, "stderr": stderr}, status_code=500)

        # Attempt to find the produced result image
        result_path = find_latest_result(run_out_dir, prefix="result")
        if not result_path:
            # fallback: try vis_result or any png
            result_path = find_latest_result(run_out_dir, prefix="vis_result") or \
                          next((os.path.join(run_out_dir, f) for f in os.listdir(run_out_dir) if f.endswith(".png")), None)

        if not result_path or not os.path.isfile(result_path):
            return JSONResponse({"error": "No result image found in output dir", "out_dir": run_out_dir}, status_code=500)

        # Read file -> base64
        with open(result_path, "rb") as f:
            data = f.read()
        img_b64 = base64.b64encode(data).decode("utf-8")

        elapsed = time.time() - t0
        print(f"[generate] Completed in {elapsed:.1f}s. result_path: {result_path}")

        return JSONResponse({"image_base64": img_b64, "result_path": result_path, "out_dir": run_out_dir})

    except Exception as e:
        import traceback
        traceback.print_exc()
        return JSONResponse({"error": str(e)}, status_code=500)

    finally:
        # NOTE: keep results for debugging; do not delete run_out_dir; but we should clean tmpdir
        try:
            shutil.rmtree(tmpdir)
        except Exception:
            pass

# -------------------------
# Optional: /sequential endpoint
# -------------------------
@app.post("/sequential")
async def sequential_apply(
    # The client provides JSON-like sequential instructions as a stringified JSON list:
    # mask_info_list = [
    #   {"mask_url": "...", "incorrect_text": "...", "correction_text": "...", "confidence": 0.9, "reason": "..."},
    #   ...
    # ]
    image_url: Optional[str] = Form(None),
    image_file: UploadFile = File(None),
    mask_info_list: str = Form(...),  # must be JSON string
    num_inference_steps: int = Form(50),
    guidance_scale: float = Form(1.0),
):
    """
    Sequentially apply corrections. This wraps multiple calls to the CLI script,
    using the previous output as the next source image.
    Returns final image_base64 and list of step outputs.
    """
    try:
        mask_infos = json.loads(mask_info_list)
    except Exception as e:
        return JSONResponse({"error": "mask_info_list must be valid JSON"}, status_code=400)

    # Prepare initial source image
    tmpdir = tempfile.mkdtemp(prefix="calligrapher_seq_")
    try:
        source_path = os.path.join(tmpdir, "source.png")
        if image_file:
            content = await image_file.read()
            with open(source_path, "wb") as f:
                f.write(content)
        elif image_url:
            if not download_to_file(image_url, source_path):
                return JSONResponse({"error": "Failed to download image_url"}, status_code=400)
        else:
            return JSONResponse({"error": "No input image provided"}, status_code=400)

        current_image_path = source_path
        step_outputs = []

        for idx, mask_info in enumerate(mask_infos, start=1):
            mask_url = mask_info.get("mask_url")
            mask_file_bytes = mask_info.get("mask_file_bytes")  # not supported here; use same upload endpoint instead
            correction_text = mask_info.get("correction_text") or mask_info.get("text") or ""
            if not mask_url:
                return JSONResponse({"error": f"mask_url missing for step {idx}"}, status_code=400)

            # Download mask for this step
            mask_path = os.path.join(tmpdir, f"mask_{idx}.png")
            if not download_to_file(mask_url, mask_path):
                return JSONResponse({"error": f"Failed to download mask for step {idx}"}, status_code=400)

            # Call CLI for this single step (self-customization, using current_image_path and mask_path)
            run_out_dir = os.path.abspath(os.path.join(DEFAULT_OUTPUT_DIR, f"seq_run_{int(time.time())}_{idx}"))
            os.makedirs(run_out_dir, exist_ok=True)

            cmd = ["python", SELF_SCRIPT,
                   "--source", current_image_path,
                   "--mask", mask_path,
                   "--prompt", correction_text,
                   "--out_dir", run_out_dir,
                   "--num_inference_steps", str(num_inference_steps),
                   "--cfg_scale", str(guidance_scale),
                   "--use_context", "1"
                   ]

            print(f"[sequential] Step {idx} cmd: {' '.join(cmd)}")
            rc, stdout, stderr = run_cli_script(cmd, cwd=os.getcwd())
            print(f"[sequential] rc={rc}; stdout length={len(stdout)}; stderr length={len(stderr)}")
            if rc != 0:
                return JSONResponse({"error": f"Step {idx} failed", "stdout": stdout, "stderr": stderr}, status_code=500)

            # Find result
            result_path = find_latest_result(run_out_dir, prefix="result") or find_latest_result(run_out_dir, prefix="vis_result")
            if not result_path:
                return JSONResponse({"error": f"No result image found for step {idx}", "out_dir": run_out_dir}, status_code=500)

            # Save as current_image_path for next iteration
            current_image_path = result_path
            with open(current_image_path, "rb") as f:
                b64 = base64.b64encode(f.read()).decode("utf-8")
            step_outputs.append({"step": idx, "result_path": result_path, "image_base64": b64})

        # Return the final image as well
        final_b64 = step_outputs[-1]["image_base64"] if step_outputs else None
        return JSONResponse({"steps": step_outputs, "final_image_base64": final_b64})

    finally:
        try:
            shutil.rmtree(tmpdir)
        except Exception:
            pass

# -------------------------
# Run with ngrok
# -------------------------
if __name__ == "__main__":
    # create output root if not exists
    os.makedirs(DEFAULT_OUTPUT_DIR, exist_ok=True)

    # start ngrok
    public_url = ngrok.connect(8000)
    print(f"🚀 Public endpoint: {public_url.public_url}/generate")
    print(f"🚀 Public endpoint (sequential): {public_url.public_url}/sequential")

    uvicorn.run(app, host="0.0.0.0", port=8000)
```
